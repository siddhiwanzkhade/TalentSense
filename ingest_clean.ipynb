### NORMALIZED_SALARY LOGIC FIXING (ANALYSED THROUGH EDA)

#### Why do we need normalized salary in TalentSpend?
Goal :
 -> Analyze hiring trends and predict salary patterns across roles, experience levels, and time.


Without normalization:
-> A 25 dollar per hr intern looks “cheap” compared to $120K/year .....(which is wrong)

-> Monthly vs yearly salaries break analysis

-> ML models learn garbage

-> Visualizations lie

=> Normalized salary is mandatory for correctness.

```python
def normalize_salary(row):
    salary = row["normalized_salary"]  # or min_salary / med_salary
    period = row["pay_period"]

    if pd.isna(salary) or pd.isna(period):
        return None

    period = period.lower()

    if period == "hourly":
        return salary * 40 * 52
    elif period == "monthly":
        return salary * 12
    elif period == "yearly":
        return salary
    else:
        return None
```

```python
df["annual_salary"] = df.apply(normalize_salary, axis=1)
```
